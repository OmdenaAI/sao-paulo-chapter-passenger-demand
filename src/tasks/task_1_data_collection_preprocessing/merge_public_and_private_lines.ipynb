```python
df_public = pd.read_csv('passenger_entrance_2022.csv')
df_public.drop(columns='Unnamed: 0', inplace=True)
df_public['month'] = pd.to_datetime(df_public['month'])
df_public
```

```python
df_merged = pd.DataFrame()
df_merged['Date']= pd.concat([df_private['year_month'],df_public['month']])
df_merged['Line Name']= pd.concat([df_private['line_color'],df_public['station']])
df_merged['Total']= pd.concat([df_private['total'],df_public['total']])
df_merged['MDU']= pd.concat([df_private['MDU (Business Days Mean)'],df_public['MDU']])
df_merged['MSA']= pd.concat([df_private['MSD (Saturdays Mean)'],df_public['MSA']])
df_merged['MDO']= pd.concat([df_private['MDO (Sundays Mean)'],df_public['MDO']])
df_merged['MAX']= pd.concat([df_private['MAX (Daily Max)'],df_public['MAX']])

df_merged.sort_values(by='Date',inplace=True)
```

```python
df_merged.to_csv('passanger_demand_public_private.csv')
```
