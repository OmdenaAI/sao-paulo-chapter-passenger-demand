```python
import pandas as pd 
```

```python
filename = "PassengerTranspoByLine\Passageiros Transportados por Linha - 2021.csv"
```

```python
cols = ['Month',
        'Total',
'Average Weekday',
'Average Saturday',
'Average Sunday',
'Daily Maximum']
```

```python
skip = 5 
start_col = 0 
end_col = 6 
combined_df = []
Line  = 1 
start_col = 0 
stop_col = 6 
```

### Creates a list of dataframes which contain Passenger travelled by a particular line 

2021

```python
while skip < 44:
    line_df = pd.read_csv(filename, delimiter=';',skiprows=skip,encoding='latin-1').iloc[:13,start_col:end_col]
    line_df.columns= cols
    line_df['Line'] = f"{Line}"
    Line += 1
    start_col += 7
    end_col += 7
    combined_df.append(line_df)
    line_df = pd.read_csv(filename, delimiter=';',skiprows=skip,encoding='latin-1').iloc[:13,start_col:end_col]
    line_df.columns= cols
    line_df['Line'] = f"{Line}"
    Line += 1
    start_col = 0
    end_col = 6
    combined_df.append(line_df)
    skip = skip + 19
    
    
```

removing the last dataframe as it is not required 

```python
combined_df.pop()
```

combining all the dataframes from the list of dataframes 

```python
PassengerBy_Line = pd.DataFrame()
for i in combined_df:
    PassengerBy_Line = PassengerBy_Line.append(i)
```

Changing line 4 to line 15 based on dataset 
and making renaming the last dataframe ( which was the dataframe with sum of all the 4 lines ) to combined 

```python
PassengerBy_Line['Line'] = PassengerBy_Line['Line'].replace(['4'],'15')
PassengerBy_Line['Line'] = PassengerBy_Line['Line'].replace(['5'],'Combined Values')

PassengerBy_Line['Year'] = '2021'
```
