```python
import pandas as pd 
```

```python
filename1 = r'Entrada de Passageiros por Estação - Média Dias Úteis - 2021.csv'
filename2 = r'Entrada de Passageiros por Estação - Média dias Úteis - 2022_0.csv'
filename3 = r'Entrada de Passageiros por Estação - Média Dias Úteis - 2023_1.csv'
```

```python
cols = ['Station','January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'Average']
```

```python
def return_df_list(filename,year):
    listDf = []
    line1 = pd.read_csv(filename,delimiter=';',skiprows=5,encoding='latin-1').iloc[:24,0:14]
    line1.columns = cols  
    line1['Line'] = '1'  
    line1['Year'] = year
    listDf.append(line1)

    line2 = pd.read_csv(filename,delimiter=';',skiprows=35,encoding='latin-1').iloc[:15,0:14]
    line2.columns = cols
    line2['Line'] = '2'
    line2['Year'] = year
    listDf.append(line2)

    line3 = pd.read_csv(filename,delimiter=';',skiprows=56,encoding='latin-1').iloc[:19,0:14]
    line3.columns = cols
    line3['Line'] = '3'
    line3['Year'] = year
    listDf.append(line3)

    line15 = pd.read_csv(filename,delimiter=';',skiprows=80,encoding='latin-1').iloc[:12,0:14]
    line15.columns = cols
    line15['Line'] = '15'
    line15['Year'] = year
    listDf.append(line15)
    return listDf
```

```python
Combined_list =[]
```

```python
listDf = return_df_list(filename1,'2021')
EntryByStation_df_21 = pd.concat(listDf)
Combined_list.append(EntryByStation_df_21)

listDf = return_df_list(filename2,'2022')
EntryByStation_df_22 = pd.concat(listDf)
Combined_list.append(EntryByStation_df_22)

listDf = return_df_list(filename3,'2023')
EntryByStation_df_23 = pd.concat(listDf)
Combined_list.append(EntryByStation_df_23)
```

```python
Passenger_EntryByStation = pd.concat([EntryByStation_df_21, EntryByStation_df_22,EntryByStation_df_23], ignore_index=True)
Passenger_EntryByStation
Passenger_EntryByStation.to_csv("Passenger_EntryByStation.csv")
Passenger_EntryByStation
```
