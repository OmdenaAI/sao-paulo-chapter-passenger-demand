```python
import pandas as pd
import numpy as np
import os
```

```python
def methodclean(filename, month, encoder):
    df_1 = pd.read_csv(filename, encoding=encoder, skiprows=5, sep=';', nrows=23, usecols=[0,1])
    df_1['Line'] = 'Azul'
    df_1['Line No'] = 1
    df_1.rename(columns={'Estação ': 'Station', 'Entradas ': 'Num of Entries'}, inplace=True)

    df_2 = pd.read_csv(filename, encoding=encoder, skiprows=5,sep=';',nrows=14, usecols=[2,3])
    df_2['Line'] = 'Verde'
    df_2['Line No'] = 2
    df_2.rename(columns={'Estação .1': 'Station', 'Entradas': 'Num of Entries'}, inplace=True)

    df_3 = pd.read_csv(filename,encoding=encoder,skiprows=5,sep=';',nrows=18,usecols=[4,5])
    df_3['Line'] = 'Vermelha'
    df_3['Line No'] = 3
    df_3.rename(columns={'Estação .2': 'Station', 'Entradas.1': 'Num of Entries'}, inplace=True)
    
    df_5 = pd.read_csv(filename,encoding=encoder,skiprows=5,sep=';',nrows=15,usecols=[6,7])
    df_5['Line'] = 'Lilás'
    df_5['Line No'] = 5
    df_5 = df_5[(df_5 != '- ').all(axis=1)]
    df_5.rename(columns={'Estação .3': 'Station', 'Entradas.2': 'Num of Entries'}, inplace=True)

    df_15 = pd.read_csv(filename, encoding=encoder, skiprows=5, sep=';',nrows=2,usecols=[8,9])
    df_15['Line'] = 'Prata'
    df_15['Line No'] = 15
    df_15.rename(columns={'Estação .4': 'Station', 'Entradas.3': 'Num of Entries'}, inplace=True)
    
    df2018 = pd.concat([df_1, df_2, df_3, df_5, df_15]).reset_index(drop=True)
    df2018['Month'] = month
    
    return df2018
```

```python
def methodclean2(filename, month, encoder):
    df_1 = pd.read_csv(filename, encoding=encoder, skiprows=5,sep=';',nrows=23, usecols=[0,1])

    df_1['Line'] = 'Azul'
    df_1['Line No'] = 1
    df_1.rename(columns={'Estação ': 'Station', 'Entradas ': 'Num of Entries'}, inplace=True)

    df_2 = pd.read_csv(filename,encoding=encoder, skiprows=5,sep=';', nrows=23, usecols=[2,3])
    df_2['Line'] = 'Verde'
    df_2['Line No'] = 2
    df_2.rename(columns={'Estação .1': 'Station', 'Entradas': 'Num of Entries'}, inplace=True)

    df_3 = pd.read_csv(filename,encoding=encoder,skiprows=5,sep=';',nrows=23, usecols=[4,5])
    df_3['Line'] = 'Vermelha'
    df_3['Line No'] = 3
    df_3.rename(columns={'Estação .2': 'Station', 'Entradas.1': 'Num of Entries'}, inplace=True)
  
    df_15 = pd.read_csv(filename, encoding=encoder,skiprows=5,sep=';',nrows=23,usecols=[6,7])
    df_15['Line'] = 'Prata'
    df_15['Line No'] = 15

    df_15.rename(columns={'Estação .3': 'Station', 'Entradas.2': 'Num of Entries'}, inplace=True)

    full_df = pd.concat([df_1, df_2, df_3, df_15]).reset_index(drop=True)
    full_df = full_df[(full_df != '- ').all(axis=1)]
    full_df['Month'] = month
    
    return full_df
```

### 2018

```python
all = []

folder_to_view = "./data/2018/entry_station"

for i, file in enumerate(sorted(os.listdir(folder_to_view))):

    if i == 0:
        continue
  
    if i < 6 or  i> 7:
        x = methodclean(folder_to_view + '/'  + file, i, 'latin_1')
        all.append(x)
    else: 
        x = methodclean(folder_to_view + '/'  + file, i, 'utf-8')
        all.append(x)

df = pd.concat(all).reset_index(drop=True)
df
```

```python
all2 = []

folder_to_view = "./data/2018/entry_station9-12"

for i, file in enumerate(sorted(os.listdir(folder_to_view)), 9):

    print(i)
    print(folder_to_view + '/'  + file)
    x = methodclean2(folder_to_view + '/'  + file, i, 'latin_1')
    all2.append(x)

df2 = pd.concat(all2).reset_index(drop=True)
df2018 = pd.concat([df, df2]).reset_index(drop=True)
df2018['Year'] = 2018
df2018
```

### 2019

```python
all3 = []

folder_to_view = "./data/Demanda - 2019/entry_station"

for i, file in enumerate(sorted(os.listdir(folder_to_view))):

    if i==0:
        continue

  
    x = methodclean2(folder_to_view + '/'  + file, i, 'latin_1')
    all3.append(x)

df2019 = pd.concat(all3).reset_index(drop=True)
df2019['Year'] = 2019
df2019
```

### 2020

```python
all4 = []

folder_to_view = "./data/2020"

for i, file in enumerate(sorted(os.listdir(folder_to_view))):

    if i ==0:
        continue
        
    x = methodclean2(folder_to_view + '/'  + file, i, 'latin_1')
    all4.append(x)

df2020 = pd.concat(all4).reset_index(drop=True)
df2020['Year'] = 2020
df2020
```
