```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from pathlib import Path
```

# Data Extraction

```python
ROOT_PATH = Path().resolve().parents[3]
ROOT_PATH
```

```python
DATA_PATH = ROOT_PATH/"src"/"tasks"/"task_1_data_collection_preprocessing"/"final_datasets"
DATA_PATH
```

# Passenger entry by station analysis

```python
pes_complete_data = [file for file in list(DATA_PATH.glob("l*_pes_complete.csv"))]

for file in pes_complete_data:
    
    print(file.name)
    df = pd.read_csv(file)
    df.index = df['month']
    #display(df.head())
    
    fig = go.Figure()
    fig.update_yaxes(title_text="dpea")
    fig.update_xaxes(title_text="month")

    for station in df['station'].unique():  

        fig.add_trace(go.Scatter(
        x=df[df["station"] == station].index,
        y=df[df["station"] == station]['dpea'],
        name=station,
        line_shape='spline'
        ))

    fig.update_layout()
    fig.show()
```
